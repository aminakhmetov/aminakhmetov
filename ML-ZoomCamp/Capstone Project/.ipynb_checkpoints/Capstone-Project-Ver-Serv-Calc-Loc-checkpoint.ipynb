## Training of Model

Separating to validation and train datasets:

```python
val_cutoff = int(len(df_train_full) * 0.8)
df_train = df_train_full[:val_cutoff]
df_val = df_train_full[val_cutoff:]
```

Now let's create image generators and set for 150x150 image size

```python
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = train_datagen.flow_from_dataframe(
    df_train,
    x_col='filename',
    y_col='label',
    target_size=(150, 150),
    batch_size=32,
)

val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

val_generator = val_datagen.flow_from_dataframe(
    df_val,
    x_col='filename',
    y_col='label',
    target_size=(150, 150),
    batch_size=32,
)
```

### Training of the base-line model

Here we are training base-line model as proposed in Kaggle competition. 

This would be the function that is setting seed in order to get reproducible results

```python
def set_seed():
    seed = 40
    np.random.seed(seed)
    tf.random.set_seed(seed)
```

The baseline model have:<br>
1. Preprocessing from [Xception model](https://keras.io/api/applications/xception/) by Convolutional Neural Network (CNN) layers.
2. Average 2D pooling.
3. Fully connected layer of 6 neurons.

```python
set_seed()
# CNN input from Xception model
base_model = Xception(
    weights='imagenet',
    include_top=False,
    input_shape=(150, 150, 3)
)
#
base_model.trainable = False

inputs = keras.Input(shape=(150, 150, 3))

base = base_model(inputs, training=False)
vectors = keras.layers.GlobalAveragePooling2D()(base)
outputs = keras.layers.Dense(6)(vectors)

model = keras.Model(inputs, outputs)
model.summary()
```

Lets set hyperparameters:

```python
# Selecting learning rate of 0.01
learning_rate = 0.01
# Setting an ADAM optimizer
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
# Setting Categorical Cross Entropy as optimization metric, because we solving classification task
loss = keras.losses.CategoricalCrossentropy(from_logits=True)
# Setting test metric -- accuracy.
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
```

Lets train model during 2 epochs

```python
history = model.fit(
    train_generator,
    epochs=2,
    validation_data=val_generator
)
```

We achieved 0.8525 of accuracy for our base-line model.

### Maxing out our knowledge on Machine Leargning: training better classification model.

Here we will implement all our knowledge and tips and tricks learngin during ML Zoomcamp or earlier:
1. First lets **try increasing image input size from 150x150 up to 300x300**, supposing that we will pass more information to the model to make a good prediction.

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
train_generator = train_datagen.flow_from_dataframe(df_train,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_generator = val_datagen.flow_from_dataframe(df_val,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32,)
set_seed()
# CNN input from Xception model
base_model = Xception(weights='imagenet',include_top=False,input_shape=(300, 300, 3))
base_model.trainable = False
inputs = keras.Input(shape=(300, 300, 3))
base = base_model(inputs, training=False)
vectors = keras.layers.GlobalAveragePooling2D()(base)
outputs = keras.layers.Dense(6)(vectors)
model = keras.Model(inputs, outputs)
model.summary()
# Selecting learning rate of 0.01
learning_rate = 0.01
# Setting an ADAM optimizer
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
# Setting Categorical Cross Entropy as optimization metric, because we solving classification task
loss = keras.losses.CategoricalCrossentropy(from_logits=True)
# Setting test metric -- accuracy.
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
history = model.fit(train_generator,epochs=2,validation_data=val_generator)
```

We get already 0.95(0.84 before) of accuracy just by increasing image input size from<br>
2. Lets **try increasing batch size**:

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
train_generator = train_datagen.flow_from_dataframe(df_train,x_col='filename',y_col='label',target_size=(300, 300),batch_size=64)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_generator = val_datagen.flow_from_dataframe(df_val,x_col='filename',y_col='label',target_size=(300, 300),batch_size=64)
set_seed()
# CNN input from Xception model
base_model_2x_batch_size = Xception(weights='imagenet',include_top=False,input_shape=(300, 300, 3))
base_model_2x_batch_size.trainable = False
inputs = keras.Input(shape=(300, 300, 3))
base = base_model_2x_batch_size(inputs, training=False)
vectors = keras.layers.GlobalAveragePooling2D()(base)
outputs = keras.layers.Dense(6)(vectors)
model_2x_batch_size = keras.Model(inputs, outputs)
model_2x_batch_size.summary()
# Selecting learning rate of 0.01
learning_rate = 0.01
# Setting an ADAM optimizer
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
# Setting Categorical Cross Entropy as optimization metric, because we solving classification task
loss = keras.losses.CategoricalCrossentropy(from_logits=True)
# Setting test metric -- accuracy.
model_2x_batch_size.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
history = model_2x_batch_size.fit(train_generator,epochs=2,validation_data=val_generator)
```

3. Now lets **try increasing number of training epochs to 10** epochs.

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
train_generator = train_datagen.flow_from_dataframe(df_train,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_generator = val_datagen.flow_from_dataframe(df_val,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
set_seed()
base_model_10_ep = Xception(weights='imagenet',include_top=False,input_shape=(300, 300, 3))
base_model_10_ep.trainable = False
inputs = keras.Input(shape=(300, 300, 3))
base = base_model_10_ep(inputs, training=False)
vectors = keras.layers.GlobalAveragePooling2D()(base)
outputs = keras.layers.Dense(6)(vectors)
model_10_ep = keras.Model(inputs, outputs)
model_10_ep.summary()
# Selecting learning rate of 0.01
learning_rate = 0.01
# Setting an ADAM optimizer
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
# Setting Categorical Cross Entropy as optimization metric, because we solving classification task
loss = keras.losses.CategoricalCrossentropy(from_logits=True)
# Setting test metric -- accuracy.
model_10_ep.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
history = model_10_ep.fit(train_generator,epochs=10,validation_data=val_generator)
```

```python
plt.plot(np.arange(1,len(history.history['val_accuracy'])+1),history.history['val_accuracy'], label='validation')
plt.plot(np.arange(1,len(history.history['val_accuracy'])+1),history.history['accuracy'], label='train')
plt.title('Accuracy vs Epochs')
plt.xticks(np.arange(1,len(history.history['val_loss'])+1),)
plt.legend()
```

```python
plt.plot(np.arange(1,len(history.history['val_loss'])+1),history.history['val_loss'], label='validation')
plt.plot(np.arange(1,len(history.history['val_loss'])+1),history.history['loss'], label='train')
plt.title('Loss vs Epochs')
plt.xticks(np.arange(1,len(history.history['val_loss'])+1),)
plt.legend()
```

We see that loss and accuracy of train and validation sets intersect between first and second epochs. That means that after two epochs we have overfiting. So to avoid overfitting we conclude that **2 epochs are better** than other value.

4. Lets **try chaning learning rate** in the [0.00001,0.0001,0.001,0.01,0.1] range

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
train_generator = train_datagen.flow_from_dataframe(df_train,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_generator = val_datagen.flow_from_dataframe(df_val,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
lr_list = [0.00001,0.0001,0.001,0.01,0.1]
history = []
for lr in lr_list:
    set_seed()
    base_model_lr = Xception(weights='imagenet',include_top=False,input_shape=(300, 300, 3))
    base_model_lr.trainable = False
    inputs = keras.Input(shape=(300, 300, 3))
    base = base_model_lr(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    outputs = keras.layers.Dense(6)(vectors)
    model_lr = keras.Model(inputs, outputs)
    # Selecting learning rate of 0.01
    learning_rate = lr
    print("Testing Learning Rate:",lr)
    # Setting an ADAM optimizer
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    # Setting Categorical Cross Entropy as optimization metric, because we solving classification task
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)
    # Setting test metric -- accuracy.
    model_lr.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    history.append(model_lr.fit(train_generator,epochs=2,validation_data=val_generator))
```

```python
for idx,lr in enumerate(lr_list):
    plt.plot(np.arange(1,len(history[idx].history['val_accuracy'])+1),history[idx].history['val_accuracy'], label='LR='+str(lr)+' :validation')
    plt.plot(np.arange(1,len(history[idx].history['val_accuracy'])+1),history[idx].history['accuracy'], label='LR='+str(lr)+' :train')
plt.title('Accuracy vs Learning Rate vs Epochs')
plt.xticks(np.arange(1,len(history[idx].history['val_loss'])+1),)
plt.xlabel('Epochs')
plt.legend()
```

```python
val_history_at_last_ep = []
train_history_at_last_ep = []
for idx,lr in enumerate(lr_list):
    val_history_at_last_ep.append(history[idx].history['val_accuracy'][-1])
    train_history_at_last_ep.append(history[idx].history['accuracy'][-1])
    
plt.plot(lr_list,val_history_at_last_ep, label='validation')
plt.plot(lr_list,train_history_at_last_ep, label='train')
plt.xscale('log')
plt.title('Accuracy vs Learning Rate @ Last Epoch')
plt.xticks(lr_list)
plt.xlabel('Learning Rate')
plt.legend()
```

We see that the best Learning Rate is 0.01

6. Lets **add a dense layer that has size of 2xCategories**

7. Lets **try to add a dropout layer** and provision x2 epochs

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
train_generator = train_datagen.flow_from_dataframe(df_train,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_generator = val_datagen.flow_from_dataframe(df_val,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
set_seed()
base_model_dl = Xception(weights='imagenet',include_top=False,input_shape=(300, 300, 3))
base_model_dl.trainable = False
inputs = keras.Input(shape=(300, 300, 3))
base = base_model_dl(inputs, training=False)
vectors = keras.layers.GlobalAveragePooling2D()(base)
drop_out = keras.layers.Dropout(0.2) (vectors)
outputs = keras.layers.Dense(6)(drop_out)
model_dl = keras.Model(inputs, outputs)
model_dl.summary()
# Selecting learning rate of 0.01
learning_rate = 0.01
# Setting an ADAM optimizer
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
# Setting Categorical Cross Entropy as optimization metric, because we solving classification task
loss = keras.losses.CategoricalCrossentropy(from_logits=True)
# Setting test metric -- accuracy.
model_dl.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
history = model_dl.fit(train_generator,epochs=4,validation_data=val_generator)
```

8. It seems that **the only best strategy that lead us to improving score -- is increasing size of figures**, lets try then more!

```python
sizes = [750,600,450,300,150]
history_sizes = []
for size in sizes:
    print("Testing Image Size:",size,'x',size)
    train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
    train_generator = train_datagen.flow_from_dataframe(df_train,x_col='filename',y_col='label',target_size=(size, size),batch_size=32)
    val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
    val_generator = val_datagen.flow_from_dataframe(df_val,x_col='filename',y_col='label',target_size=(size, size),batch_size=32)
    set_seed()
    base_model_lr = Xception(weights='imagenet',include_top=False,input_shape=(size, size, 3))
    base_model_lr.trainable = False
    inputs = keras.Input(shape=(size, size, 3))
    base = base_model_lr(inputs, training=False)
    vectors = keras.layers.GlobalAveragePooling2D()(base)
    outputs = keras.layers.Dense(6)(vectors)
    model_lr = keras.Model(inputs, outputs)
    # Selecting learning rate of 0.01
    learning_rate = 0.01
    # Setting an ADAM optimizer
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    # Setting Categorical Cross Entropy as optimization metric, because we solving classification task
    loss = keras.losses.CategoricalCrossentropy(from_logits=True)
    # Setting test metric -- accuracy.
    model_lr.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    history_sizes.append(model_lr.fit(train_generator,epochs=2,validation_data=val_generator))
```

```python
val_history_at_last_ep = []
train_history_at_last_ep = []
for idx,size in enumerate(sizes):
    val_history_at_last_ep.append(history_sizes[idx].history['val_accuracy'][-1])
    train_history_at_last_ep.append(history_sizes[idx].history['accuracy'][-1])
plt.plot(sizes,val_history_at_last_ep, label='validation')
plt.plot(sizes,train_history_at_last_ep, label='train')
plt.title('Accuracy vs Image Sizes @ Last Epoch')
plt.xticks(sizes)
plt.xlabel('Image Sizes')
plt.legend()
```

We have next validation accuracy scores:
1. For 750x750: 0.9577<br>
2. For 600x600: 0.9523<br>
3. For 450x450: 0.9568<br>
4. For 300x300: 0.9371<br>
5. For 150x150: 0.8624<br>
It seems that image size 450x450 gives best score of **~0.9568** on validation data. However, earlier at step 1 we got **0.9550** for 300x300, so in order to save in model complexity we are choosing final solution 300x300, that has very close perofrmance to 450x450

9. Lets train final model with all adjusted parameters  on full data

```python
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
train_generator = train_datagen.flow_from_dataframe(df_train_full,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_generator = val_datagen.flow_from_dataframe(df_val,x_col='filename',y_col='label',target_size=(300, 300),batch_size=32)
base_model_full = Xception(weights='imagenet',include_top=False,input_shape=(300, 300, 3))
base_model_full.trainable = False
inputs = keras.Input(shape=(300, 300, 3))
set_seed()
base = base_model_full(inputs, training=False)
vectors = keras.layers.GlobalAveragePooling2D()(base)
outputs = keras.layers.Dense(6)(vectors)
model_full = keras.Model(inputs, outputs)
model_full.summary()
# Selecting learning rate of 0.01
learning_rate = 0.01
# Setting an ADAM optimizer
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
# Setting Categorical Cross Entropy as optimization metric, because we solving classification task
loss = keras.losses.CategoricalCrossentropy(from_logits=True)
# Setting test metric -- accuracy.
model_full.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
history = model_full.fit(train_generator,epochs=2,validation_data=val_generator)
```

```python
# saving model
model_full.save("cutlery-loc.h5")
```
